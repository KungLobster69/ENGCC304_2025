```python
score_list = [88, 92, 75, 68, 95, 80, 50, 40, 30]

def calculate_average(score_list):
    if len(score_list) == 0:
        return 0
    
    total = sum(score_list)
    average = total / len(score_list)
    return average


def find_max(score_list):
    if len(score_list) == 0:
        return None
    
    max_score = score_list[0]
    for score in score_list:
        if score > max_score:
            max_score = score
    return max_score


def find_min(score_list):
    if len(score_list) == 0:
        return None
    
    min_score = score_list[0]
    for score in score_list:
        if score < min_score:
            min_score = score
    return min_score


def count_passed(score_list):
    student_count = 0
    for score in score_list:
        if score >= 50:
            student_count += 1
    return student_count


def get_grade(score_list):
    total_grade = []
    for score in score_list:
        if score >= 80:
            grade = 'A'
        elif score >= 70:
            grade = 'B'
        elif score >= 60:
            grade = 'C'
        elif score >= 50:
            grade = 'D'
        else:
            grade = 'F'
        total_grade.append(grade)
    return total_grade



avg = calculate_average(score_list)
heighest = find_max(score_list)
lowest = find_min(score_list)
student_passed = count_passed(score_list)
student_grade = get_grade(score_list)

print('--- Score Analysis ---')
print(f'คะแนนทั้งหมด: {score_list}')
print(f'คะแนนสูงสุด: {heighest}')
print(f'คะแนนต่ำสุด: {lowest}')
print(f'คะแนนเฉลี่ย: {avg:.2f}')
print(f'=='* 20)
print(f'จำนวนนักเรียนที่สอบผ่าน: {student_passed}')
print(f'เกรดของนักเรียนแต่ละคน: {student_grade}')
```

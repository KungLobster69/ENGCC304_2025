```python
import random

#สุ่มตัวเลขระหว่าง 1 ถึง 100
answer = random.randint(1, 100)

print("ยินดีต้อนรับสู่เกมทายตัวเลข!")
print("ฉันได้สุ่มตัวเลขระหว่าง 1 ถึง 100 แล้ว ลองทายดูสิ!")

while True:
    try:
        # รับตัวเลขที่ผู้เล่นทาย
        guess = int(input("ป้อนตัวเลขที่ทาย: "))

        # เปรียบเทียบและให้คำใบ้
        if guess == answer:
            print("คุณทายถูกแล้ว!")
            break  # ออกจาก Loop เมื่อทายถูก
        elif guess < answer:
            print("น้อยเกินไป!")
        else:
            print("มากเกินไป!")
    except ValueError:
        print("ป้อนค่าไม่ถูกต้อง กรุณาป้อนตัวเลขจำนวนเต็ม.")
```
